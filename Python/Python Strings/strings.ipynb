### Creating Strings

```python
s = 'Hello'
s = "Hello"

# Multiline String
s = '''Hello'''
s = """Hello"""

# Using Type Conversion
s = str('Hello')

print(s)

# Options are given to avoid problems like 
# 'It's raining outside'
```

### Accessing Substrings from a String

```python
# Positive Indexing
s = 'Hello World'
print(s[4])
# 0 t0 (len - 1) in total indexes possible
# print(s[41]) = error
```

```python
# Neegative Indexing
s = 'Hello World'
# we will go reverse direction start from -1 
# move from right to left
print(s[-2])

# benefit - without even knowing the length of the string we can do indexing
```

```python
# Slicing - we can extract more than one characters
s = 'Hello World'
print(s[2:3])
print(s[2:]) # 2 to complete end
print(s[:3]) # 0 to 3rd character
print(s[:]) # print entire string
print(s[0:6:2]) # with step size of 2
```

```python

s = "hello World"
print(s[6:0:-2]) # in negative indexing first number should be greater than 2nd number
```

```python
s = "hello World"
# to reverse any string
print(s[::-1])
```

```python
# Slicing using negative indexing
s = 'hello World'
print(s[-5:-1])
print(s[-5:])
print(s[-1:-6:-1])
```

## Example Programs

```python
# Find the length of a given string without using the len() function

s = input('Enter the string:-')
counter = 0

for i in s:
    counter += 1
print(counter)
```

```python
# Extract username from a given email.
# Eg if the email is nitish24singh@gmail.com
# then the username should be nitish24singh.

s = input('Enter the email:-')

pos = s.index('@')
print(s[0:pos])
```

```python
#Count the frequency of a particular character in a provided string.
# Eg 'hello how are you' is the string,the frequency of h in this string is 2.

n = input('enter the string:-')
p = input('Enter the character :-')
frequency = n.count(p)

print('the frequency of ',p,'in this string is',frequency)
```

```python
# without built in function

n = input('enter the string:-')
p = input('Enter the character :-')

frequency = 0
for i in n:
    if i == p :
        frequency += 1

print('the frequency of ',p,'in this string is',frequency)
```

```python
# Write a program which can remove a particular character from a string.
s = input('Enter the string:-')
term = input('What would like to remove:-')

result = ''

for i in s:
    if i != term:
        result = result + i

print(result)
```

```python
# Write a program that can check whether a given string is palindrome or not.

s = input('Enter the string:-')
flag=  True
for i in range(0,len(s)//2):
    if s[i] != s[len(s)-i-1] :
        flag = False
        print('Not a palindrome')
        break

if flag :
    print('Palindrome')
```

```python
# Write a program to count the number of words in a string without split()
s = input("Enter the string:-")
count = 0
for i in s:
    if i == " " or i == ".":
        count = count + 1
print(count)
```

```python
s = input("Enter the string:-")
temp = ' '
L = []
for i in s:
    if i!=' ':
        temp = temp + i
    else:
        L.append(temp)
        temp = ''
L.append(temp)
print(L)
```

```python
# Write a python program to convert a string to title case without using the title().
s = input("enter the string:-")
L = []
for i in s.split():
    L.append(i[0].upper() + i[1:].lower())
print(" ".join(L))
```

```python
# Write a program that can convert an integer to string.
number = int(input("Enter the number:-"))
digits = '0123456789'
result = ''
while number != 0:
    result = digits[number % 10] + result
    number = number//10
print(result)
print(type(result))
```
