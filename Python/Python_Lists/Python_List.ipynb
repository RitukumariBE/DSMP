# List in Python
List is a data type where you can store multiple items under 1 name.More technically,lists act like dynamic arrays
which means you can add more items on the fly.

## Arrays Vs Lists
- Fixed Vs Dynamic Size
- Convenience -> Hetrogeneous
- Speed of Execution
- Memory

```python
L = [1,2,3]
print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[-2]))
print(id(1))
print(id(2))
print(id(3))
```

## Characteristics of a List
- Ordered
- Changable/Mutable
- Hetrogeneous
- Can have duplicates
- are dynamic
- can be nested
- items can be accessed
- can contain any kind of objects in Python


```python
L1 = [1,2,3]
L2 = [3,2,1]
L1 == L2
```

## Creating a List


```python
# Empty
print([])
# 1D
print([1,2,3,4,5])
# 2D
print([1,2,3,[4,5]])
# 3D
print([[[1,2],[3,4]],[[5,6],[7,8]]])
#Heterogeneous
print([1,True,5.6,5+6j,'Hello'])
# type conversion
print(list('Hello'))
```

## Accessing Items from a List

```python
# Indexinng
L = [1,2,3,4,5]
# positive
print(L[0])
print(L[30])
#negative
print(L[-5])
print(L[-3])

L = [1,2,3,[4,5]]
print(L[-1])
print(L[-1][-2])
print(L[3][0])

L = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(L[1][0][0])
print(L[0][0][1])
```

```python
# Slicing
L = [1,2,3,4,5]
print(L[0:3])
print(L[-3:-1])
print(L[::-1])
```
