### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
head = int(input("Enter the number of head:-")) # n
legs = int(input("Enter number of legs:-")) # m
y = (4*head - legs)/2 # number of chicken
x = head - y # number of dog

print(f"The number of dogs is {x} and number of chicken is {y}")


````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# Write your code here
n = int(input("Enter the n upto where you want to sum:-"))
sum = 0
for i in range(1,n+1):
    sum = sum + i**2
print(f"The sum of squares of {n} natural number is {sum}")
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# Write your code here
a1 = int(input("Enter the 1st term:-"))
a2 = int(input("Enter the second term:-"))
d = a2 - a1
n = int(input("Enter which term you want to find:-"))
an = a1 + ((n - 1)*d)
print(f"The {n}th term is {an}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
import math

# Tank dimensions
H = float(input("Enter height of tank (cm): "))
L = float(input("Enter length of tank (cm): "))
B = float(input("Enter breadth of tank (cm): "))

# Glass dimensions
h = float(input("Enter height of glass (cm): "))
r = float(input("Enter radius of glass (cm): "))

# Volume of tank (cuboid)
tank_volume = H * L * B

# Volume of glass (cylinder)
glass_volume = math.pi * r * r * h

# Number of glasses
num_glasses = tank_volume / glass_volume

print("Number of glasses that can be filled:", int(num_glasses))
````
