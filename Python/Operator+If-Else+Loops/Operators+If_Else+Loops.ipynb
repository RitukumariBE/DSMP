# Operators in Python
- Arithmetic Operators
- Logical Operators
- Relational Operators
- Bitwise Operation
- Assignment Operators
- Membership Operators

```python
# Arithmetic Operators
print(5 + 6 )
print(5 - 6)
print(5 * 6)
print(5 / 6)
print(6 // 5) # Integer division -> quotient ko intege me convert kr dega
print(5 % 2) # Modulas operator -> gives remainder
print(5 ** 2) # power of
```

```python
# Relational Operator

print(4 > 5)
print(4 < 5)
print(4 >= 4)
print(4 <= 4)
print(4 == 4)
print(4 != 4)
```

```python
# Logical operator
print(1 and 0) # and
print( 1 or 0) #or
print( not 1) #not
```

```python
# Bitwise Operator (Operate on binary value)

#bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# bitwise not
print(~3)

# left shift
print( 4>> 2)

# Right shift
print(5 << 2)
```

```python
# Assignment Operator
# = # a = 2
a = 2
# a = a + 2
a += 2 # we can use any arithmetic operation(+,-,/,*,%,...)
print(a)
# a++ ++a -> not in python
```

```python
# Membership Operator
# in
print('D' in 'Delhi')

# not in
print('D' not in 'Delhi')
print(1 in [2,3,4,5,6])
```

## Loops in Python
- Need for loops
- while loops
- For loops

```python
number = int(input('Enter the number:-'))
i = 1
while i < 11:
  print(number,'*',i,'=',number*i)
  i += 1
```

```python
# while loops with else
x = 1
while x < 3:
  print(x)
  x += 1

else:
  print('Limit crossed')
```

```python
# Guessing Game
import random
jackpot = random.randint(1,100)
guess = int(input('Guess karo :'))
counter = 1
while guess != jackpot:
  if guess < jackpot:
    print('Galat! Guess Higher')
  else :
    print('Galat! Guess Lower')
  guess = int(input('Guess karo:'))
  counter += 1
else:
  print('Correct Guess')
  print('Attempts',counter)
```

```python
# Sequence sum
# 1/1! + 2/2! + 3/3! + .....
n = int(input('Enter n:'))
result = 0
fact = 1
for i in range (1,n+1):
  fact = fact * i
  result = result + (i/fact)
print(result)
```

```python
# Nested Loops
for i in range(1,5):
  for j in range (1,5):
    print(i,j)
```

```python
# Pattern 1
# *
# * *
# * * *

rows = int(input("Enter number of rows:"))
for i in range (1,rows+1):
  for i in range(1,i+1):
    print('*',end='')
  print()
```

```python
# Pattern 2
# 1
# 1 2 1
# 1 2 3 2 1
# 1 2 3 4 3 2 1

rows = int(input('Enter number of rows:'))
for i in range(1,rows + 1):
  for j in range(1,i + 1):
    print(j,end = ' ')
  for k in range (i-1,0,-1):
    print(k,end = ' ')
  print()
```

```python
# Loop Control Statement
# - Break
# - Continue
# - Pass

for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input('Enter lower range:'))
upper = int(input('Enter upper range:'))
for i in range(lower,upper + 1):
  for j in range(2,i):
    if i % j == 0:
      break
    else :
      print(i)
```

```python
# continue
for i in range(1,10):
  if i== 5:
    continue
  print(i)
```

```python
#pass
for i in range(1,10):
  pass
```
