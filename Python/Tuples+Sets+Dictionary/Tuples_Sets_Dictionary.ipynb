# Tuples
A tuple in Python is similiar to a list.The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.
In short ,a tuple is an immputable list .A tuple can not be changed in any way once it is created.

Characteristics:-
- Ordered
- Unchangeble
- Allows duplicate

```python
# Creating Tuples
#empty
t1 = ()
print(t1)
#create a tuple with a single item
t2 = ('hello',)
print(type(t2))
#homo
t3 = (1,2,3,4)
print(t3)
#hetro
t4 = (1,2.5,True,[1,2,3])
print(t4)
# tuple
t5 = (1,2,3,(4,5))
print(t5)
# using type conversion
t6 = tuple('hello')
print(t6)
```

```python
# Accessing Items
# Indexing
t3 = (1,2,3,4)
print(t3)
print(t3[0])
print(t3[-1])

# Slicing
print(t3[0:4:2])
print(t3[-3:-1])
print(t3[::-1])

t5 = (1,2,3,(4,5))
t5[-1][0]
```

```python
# Editing Items
print(t3)
t3[0] = 100

# Immutable just like strings
```

```python
# Adding Items
print(t3)
# Not possible
```

```python
# Deleting Items
print(t3)
del t3
print(t3)

# tuple does not support item deletion
```

```python
# Operation on tuples
# + and * 
# membership
# iteration

t1 = (1,2,3,4)
t2 = (5,6,7,8)
print(t1 + t2)
print(t1 * 3)

print(1 in t1)

for i in t1:
    print(i)
```

```python
# Tuple Functions
# len/sum/min/max/sorted
t = (1,2,3,4)
len(t)
sum(t)
min(t)
max(t)
sorted(t)
sorted(t,reverse = True)
```

```python
# count
t = (1,2,3,4,5)
t.count(5)
t.count(50)
```

```python
# index
t.index(50)
# x not in the tuple
```

### Difference between lists and Tuples
- Syntax
- Mutability
- Speed
- Memory
- Built in functionality
- Error prone
- Usability

```python
import time
L = list(range(100000000))
T = tuple(range(100000000))
start = time.time()
for i in L:
    i*5
print('List time',time.time()-start)
start = time.time()
for i in T:
    i*5
print('Tuple time',time.time()-start)
```

```python
import sys
L = list(range(1000))
T = tuple(range(1000))
print('List Size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

```python
a = [1,2,3,4]
b = a
a.append(4)
print(a)
print(b)
```

```python
a = (1,2,3)
b = a
a = a + (4,)
print(a)
print(b)
# List me changes ho skte h whereas tuple me nhi ho skte
```

```python
# Special Unpacking
# tuple unpacking
a,b,c = (1,2,3)

a,b = (1,2,3)
print(a,b)

# valve swap
a = 1
b = 2
a,b = b,a
print(a,b)
```

```python
a,b,*others = (1,2,3,4)
print(a,b)
print(others)
```

```python
# zipping tuples
a = (1,2,3,4)
b = (5,6,7,8)
zip(a,b)
list(zip(a,b))
tuple(zip(a,b))
```

# Sets
A set is an unorderedc collection of items.Every set element is unique(no duplicates) and must be immutable(cannot be changed).However,a set itself is mutable.We can add or remove items from it.
Set can also be used to perform mathematical set operations like union,intersection,symmetric difference etc.
## Characteristics :-
- Unordered
- Mutable
- No duplicates
- Can't contain mutable data types


```python
# Creating Set
#empty
s = set()
print(s)
print(type(s))

# 1D and 2D
s1 = {1,2,3}
print(s1)
# s2 = {1,2,3,{4,5}} 
# print(s2) -> error

#hetero
s3 = {1,'hello',4.5,True,(1,2,3)}
print(s3)
s3 = {1,'hello',4.5,True,(1,2,3)}

#using type conversion
s4 = set([1,2,3])
print(s4)

#duplicates not allowed
s5 = {1,2,3,1,2,3,3}
print(s5)

# set can't have mutable items
s6 = {1,2,[3,4]}
print(s6)

#unordered
s1 = {1,2,3}
s2 = {3,2,1}
print(s1 == s2)
```
